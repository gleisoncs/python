```python
import numpy as np 

# Dados
X = np.array([8, 6, 15, 20, 25, 11, 5, 32, 28, 20])
Y = np.array([15, 20, 20, 40, 50, 25, 10, 55, 35, 30])

# Cálculos
mean_X = np.mean(X)
mean_Y = np.mean(Y)

# Coeficientes da linha de regressão
b = np.sum((X - mean_X) * (Y - mean_Y)) / np.sum((X - mean_X) ** 2)
a = mean_Y - b * mean_X

# Valores preditos (Y estimado)
Y_hat = a + b * X

# Cálculo do R-quadrado (R^2)
SS_total = np.sum((Y - mean_Y) ** 2)  # Soma dos Quadrados Totais (variabilidade total)
SS_residual = np.sum((Y - Y_hat) ** 2)  # Soma dos Quadrados dos Resíduos (variabilidade não explicada)
R_squared = 1 - (SS_residual / SS_total)  # Coeficiente de determinação

b, a, R_squared
```

```python
# Equação da linha de regressão
Y_pred = a + b * X

Y_pred
```

```python
import matplotlib.pyplot as plt

# Gráfico
plt.scatter(X, Y, color='blue', label='Dados')
plt.plot(X, Y_pred, color='red', label=f'Regressão Linear\nY = {a:.2f} + {b:.2f}X')
plt.xlabel('Distância (X)')
plt.ylabel('Tempo (Y)')
plt.legend()
plt.title('Gráfico de Dispersão com Linha de Regressão Linear')
plt.grid(True)
plt.show()
```
